```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sp # libreria para computación científica. Se requiere para leer archivo .mat
```

## Carga del registro

```python
mat_data = sp.loadmat('datos_lineales_error_y.mat')
x = mat_data['x'].flatten() # extrae datos del arreglo en una dimension
y = mat_data['y'].flatten()
```

## Salida gráfica
### Parámetros para guardar en .pdf

```python
plt.figure()
tipo_fuente = 'Arial'
tamano_fuente = 12 
plt.rc('font', family=tipo_fuente, size=tamano_fuente)

plt.plot(x, y, 'ro', linewidth=2, markersize=7)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.savefig('Datos_Modelo_Lineal.pdf', format='pdf', dpi=300)

plt.show()
```

## Regresión lineal de los parámetros $a$ y $b$
### Preparación de los datos

```python
A = np.column_stack((x, np.ones_like(x))) # Alistamiento de la matriz para regresión
```

### Regresión lineal por mínimos cuadrados
### pendiente de verificar si np.linalg.lstsq está basado a una factorización LU y si equivale a $\phi = (A^TA)^{-1}A^Ty$

```python
phi = np.linalg.lstsq(A, y, rcond=None)[0] # regresion lineal 
a, b = phi[0], phi[1]
```

### Modelo ajustado

```python
yc = a * x + b

plt.figure()
plt.plot(x, y, 'ro', label='Datos originales')
plt.plot(x, yc, 'b', label='Ajuste lineal')
plt.grid(True)
plt.xlabel('$x$', fontsize=tamano_fuente)
plt.ylabel('$y$', fontsize=tamano_fuente)
plt.title(f'Parámetros ajustados, a = {a:.4f}, b = {b:.4f}')
plt.legend()

plt.show()
```
